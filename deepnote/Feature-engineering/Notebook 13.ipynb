# Hotel Reviews

```
import pandas as pd
import numpy as np
import seaborn as sns
from textblob import TextBlob
```

```
data = pd.read_csv('Hotel_Reviews.csv')
data.head()
```

```
data['Rating'].value_counts()
```

```
sns.distplot(data['Rating'])
```

```
# Calculate the length of the reviews
data['length'] = data['Review'].apply(len)
sns.distplot(data['length'])
```

```
data['word_count'] = data['Review'].apply(lambda x: len(x.split()))
data['word_density'] = data['length'] / (data['word_count'] + 1)
```

```
sns.distplot(data['word_count'])
```

```
sns.distplot(data['word_density'])
```

```
def polar_check(x):
    textblob = TextBlob(x)
    pol = textblob.sentiment.polarity
    return pol

data['polarity'] = data['Review'].apply(polar_check)
```

```
sns.distplot(data['polarity'])
```

```
def subj_check(x):
    textblob = TextBlob(x)
    sub = textblob.sentiment.subjectivity
    return sub

data['subjectivity'] = data['Review'].apply(subj_check)
```

```
sns.distplot(data['subjectivity'])
```
