## at_time function

```
# shows 7 periods of 12 hours
i = pd.date_range('1/1/2024',periods=7,freq='12H')

ds = pd.DataFrame({'A':[40,25,34,35,36,37,38]},index=i)

ds
```

## between_time function

The `between_time` function in pandas is used to select rows of a DataFrame whose index values fall between two specified times. This function is particularly useful when working with time series data.

Parameters
 * start_time  :  The start time as a string in the format 'HH:MM' or a datetime.time object.
 * end_time  :  The end time as a string in the format 'HH:MM' or a datetime.time object.
 * include_start  :  Boolean, default True. Whether to include the start time in the result.
 * include_end  :  Boolean, default True. Whether to include the end time in the result.
 * axis  :  Axis along which to filter. By default, it is the index (0 or 'index').

```
bt = pd.date_range('1/1/2024',periods=7,freq='1D20min')

btds = pd.DataFrame({'A':[40,25,34,35,36,37,38]},index=bt)

btds
```

```
btds.between_time("00:00","00:50")
```
