# Target Encoding

```
import pandas as pd
import numpy as np
```

```
data = pd.read_csv("bigmart.csv")
data
```

```
x = data["Item_Outlet_Sales"].groupby(data["Item_Type"]).agg('mean')
x
```

```
y = np.sort(data["Item_Type"].unique())
y
```

```
x.values
```

```
data["Item_Type"] = data["Item_Type"].replace(y,x.values)
```
