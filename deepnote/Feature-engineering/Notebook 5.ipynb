# Aggregating features

```
import pandas as pd
```

```
data = pd.read_csv("Student.csv")
```

```
data.head()
```

```
data['total_score'] = data['math score'] + data['reading score'] + data['writing score']
```

```
data.head()
```

```
data["percentage"] = (data['total_score'] / 300) * 100
```

```
data.head()
```

```
def grade(x):
    if x >= 80:
        return "A"
    if x >= 70:
        return "B"
    if x >= 60:
        return "C"
    if x >= 50:
        return "D"
    elif x >= 40:
        return "E"
    else:
        return "F"

data['Grade'] = data['percentage'].apply(grade)

data.head()
```
