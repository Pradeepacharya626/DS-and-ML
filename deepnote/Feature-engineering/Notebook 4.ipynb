# Binning Numerical features

```
import pandas as pd
```

```
data = pd.read_csv("employee.csv")
data.head()
```

```
data["Age"].describe()
```

```
data["Age"].nunique()
```

```
def age_cat(x):
    if x>=18 and x<30:
        return "18-30"
    elif x>=30 and x<40:
        return "30-40"
    elif x>=40 and x<50:
        return "40-50"
    else:
        return "50-60"

data["Age"] = data["Age"].apply(age_cat)
data["Age"].value_counts()
```

```
data['DistanceFromHome'].describe()
```

```
def dist_cat(x):
    if x<=2:
        return "Near"
    elif x > 2 and x <= 5:
        return "Far"
    else:
        return "Extremly far"

data["DistanceFromHome"] = data["DistanceFromHome"].apply(dist_cat)
data["DistanceFromHome"].value_counts()
```

```
def experience_category(x):
    if x<=2:
        return "novice"
    elif x > 2 and x <= 5:
        return "intermediate"
    else:
        return "expert"

data["YearsAtCompany"] = data["YearsAtCompany"].apply(experience_category)
data["YearsAtCompany"].value_counts()
```
