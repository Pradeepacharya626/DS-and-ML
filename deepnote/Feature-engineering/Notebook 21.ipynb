# square and Cube root transformation

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data = pd.read_csv("bigmart.csv")
data.head()
```

## QQ plot

quantile-quantile plot

```
# make the qq plot for the Item_outlet_sales to check its distribution and skewness

from scipy import stats
from scipy.stats import norm

sns.distplot(data["Item_Outlet_Sales"],fit=norm)

# Get the fitted parameters used by the function
(mu,sigma) = norm.fit(data["Item_Outlet_Sales"])
print("mean = {:.2f} and standard daviation = {:.2f}".format(mu,sigma))

plt.legend(["Normal dist.($\mu=${:.2f} and $\sigma=${:.2f})".format(mu,sigma)],loc='best')
plt.ylabel("Frequency")
plt.title("Item Outlet sales distribution")

# Create a new figure
fig = plt.figure()
res = stats.probplot(data["Item_Outlet_Sales"],plot = plt)
plt.show()
```

Straight Line: If the points on the Q-Q plot lie on or close to a straight line, it suggests that the data follows the theoretical distribution (usually normal).
Deviation from Line: If the points deviate from the straight line, especially at the ends, it suggests that the data may not follow the theoretical distribution.
Slope and Intercept: The slope and intercept of the best-fit line give an indication of the scale and location of the data relative to the theoretical distribution.

### square root transformation

```
# Using the square root transformation
data["Item_Outlet_Sales"] = np.sqrt(data["Item_Outlet_Sales"])
```

```
# Again checking after sqrt transformation

sns.distplot(data["Item_Outlet_Sales"],fit=norm)

# Get the fitted parameters used by the function
(mu,sigma) = norm.fit(data["Item_Outlet_Sales"])
print("mean = {:.2f} and standard daviation = {:.2f}".format(mu,sigma))

plt.legend(["Normal dist.($\mu=${:.2f} and $\sigma=${:.2f})".format(mu,sigma)],loc='best')
plt.ylabel("Frequency")
plt.title("Item Outlet sales distribution")

# Create a new figure
fig = plt.figure()
res = stats.probplot(data["Item_Outlet_Sales"],plot = plt)
plt.show()
```

### cube root transformation

```
# Using the cube root transformation
data["Item_Outlet_Sales"] = np.cbrt(data["Item_Outlet_Sales"])
```

```
# Again checking after cbrt transformation

sns.distplot(data["Item_Outlet_Sales"],fit=norm)

# Get the fitted parameters used by the function
(mu,sigma) = norm.fit(data["Item_Outlet_Sales"])
print("mean = {:.2f} and standard daviation = {:.2f}".format(mu,sigma))

plt.legend(["Normal dist.($\mu=${:.2f} and $\sigma=${:.2f})".format(mu,sigma)],loc='best')
plt.ylabel("Frequency")
plt.title("Item Outlet sales distribution")

# Create a new figure
fig = plt.figure()
res = stats.probplot(data["Item_Outlet_Sales"],plot = plt)
plt.show()
```

### Box Cox transformation

```
data = pd.read_csv("bigmart.csv")
```

```
# Finding the skewed features
from scipy.stats import skew

numerical_feats = data.dtypes[data.dtypes != 'object'].index

# Checking the skewness of the all the numerical features
skewed_feats = data[numerical_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)

# Converting the features into dataframe
skewness = pd.DataFrame({'skew':skewed_feats})
skewness.head()
```
