# Decomposition of datetime

Date variables are special type of categorical variables

we have to extract

- year

- month

- date

- day of week

- time difference

```
import pandas as pd
import datetime
```

```
time_series = pd.date_range('1/1/2024', periods=10)
data = pd.DataFrame({"date": time_series})
```

```
data
```

```
data['date'] = pd.to_datetime(data['date'])
```

```
data['year'] = data['date'].dt.year

data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['day_of_week'] = data['date'].dt.day_of_week
data['date_only'] = data['date'].dt.date # Renamed column to 'date_only' to avoid confusion
data
```

```
df = data.select_dtypes('object')
df
```
