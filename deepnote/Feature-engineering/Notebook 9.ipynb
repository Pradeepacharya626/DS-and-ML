## n largest function

The `nlargest` function in pandas is used to return the first `n` rows of a DataFrame or Series ordered by columns in descending order. This function is particularly useful when you want to quickly find the top `n` largest values in a DataFrame or Series.


### Parameters
- `n`: int
  - Number of items to retrieve.
  -Default it is 5
- `keep`: {‘first’, ‘last’, ‘all’}, default ‘first’
  - Where there are duplicate values:
    - `first`: prioritize the first occurrence(s)
    - `last`: prioritize the last occurrence(s)
    - `all`: keep all occurrences


```
countries_population = {
    "China": 1409517397,
    "India": 1339180127,
    "USA": 324459463,
    "Indonesia": 263991379,
    "Brazil": 212559417,
    "Pakistan": 220892331,
    "Nigeria": 206139587,
    "Bangladesh": 206139587,
    "Russia": 145171738,
    "Mexico": 128932753,}

s = pd.Series(countries_population)
```

```
s.nlargest()
```

```
s.nlargest(3)
```

```
s.nlargest(7)
```

```
s.nlargest(7, keep='last') # interchange of bangladesh and nigeria takes place
```

```
# it also includes duplicate values
s.nlargest(7, keep='all')
```

```
s.nlargest(10)
```
