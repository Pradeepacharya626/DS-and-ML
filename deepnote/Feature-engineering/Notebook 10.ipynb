```
import pandas as pd
import numpy as np
```

```
df = pd.DataFrame({'A': [0,1,2,np.nan,4,5,8,np.nan]})
df
```

```
df.expanding(2).sum()
```

```
df.expanding(3).sum()
```

```
df.expanding(3).mean()
```

### Cumulative functions

```
df.cummin()
```

```
df.cummax()
```

```
df.cumsum()
```

```
df.cumprod()
```
