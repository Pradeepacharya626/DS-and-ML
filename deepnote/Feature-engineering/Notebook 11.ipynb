# Employee data feature engineering

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

here the attrition means a employee is leaving the job or not

```
data = pd.read_csv("employee.csv")
data.head()
```

> what does relationship satisfaction means

Relationship Satisfaction in the context of employee data typically refers to how satisfied an employee is with their relationships at work. This can include relationships with colleagues, supervisors, and subordinates. It is usually measured on a scale, with higher values indicating greater satisfaction. 

Let's check the unique values and their counts in the "RelationshipSatisfaction" column to understand its

```
data['RelationshipSatisfaction'].value_counts()
```

The "RelationshipSatisfaction" column has values ranging from 1 to 4, where:
- 1 indicates low satisfaction
- 2 indicates medium-low satisfaction
- 3 indicates medium-high satisfaction
- 4 indicates high satisfaction

> what is job satisfaction

Job Satisfaction in the context of employee data typically refers to how content an employee is with their job. This can include various aspects such as the nature of the work, work environment, compensation, growth opportunities, and relationships with colleagues and supervisors. It is usually measured on a scale, with higher values indicating greater satisfaction.

Let's check the unique values and their counts in the "JobSatisfaction" column to understand its distribution

```
data['JobSatisfaction'].value_counts()
```

The "JobSatisfaction" column has values ranging from 1 to 4, where:
- 1 indicates low satisfaction
- 2 indicates medium-low satisfaction
- 3 indicates medium-high satisfaction
- 4 indicates high satisfaction

```
data["overall_satisfaction"] = data["RelationshipSatisfaction"] + data["JobSatisfaction"] + data["EnvironmentSatisfaction"]
```

```
sns.distplot(data["overall_satisfaction"])
```

```
# Average time span of an employee in the company
data['AvgTimespanAtCompany'] = data['TotalWorkingYears'] / (data['NumCompaniesWorked'] +1)
```

```
sns.distplot(data["AvgTimespanAtCompany"])
```

```
def dist_cat(x):
    if x<=2:
        return "Near"
    elif x > 2 and x <= 5:
        return "Far"
    else:
        return "Extremly far"

data["DistanceFromHome"] = data["DistanceFromHome"].apply(dist_cat)
data["DistanceFromHome"].value_counts()
```

```
def age_cat(x):
    if x<40:
        return "Young"
    else:
        return "Old"

data["Age"] = data["Age"].apply(age_cat)
```
