# Decomposing categorical features

```
import pandas as pd
```

```
data = pd.read_csv('fifa.csv')
data.head()
```

```
pd.set_option('display.max_rows', 164)
data["Nationality"].value_counts()
```

```
def decompose_categorical(x):
    if x == 'England':
        return 'England'
    elif x == 'Germany':
        return 'Germany'
    elif x == 'Spain':
        return 'Spain'
    elif x == 'Argentina':
        return 'Argentina'
    else :
        return 'Others'

data["Nationality"] = data["Nationality"].apply(decompose_categorical)

data["Nationality"].value_counts()
```
