# Sunburst charts

A sunburst chart is a type of data visualization that displays hierarchical data using concentric circles. Each level of the hierarchy is represented by one ring or circle, with the innermost circle representing the top of the hierarchy. The segments within each ring represent the breakdown of the parent segment, and the size of each segment is proportional to its value.

- labels : sets the labels of sunburst sectors

- parents : sets the parent sector of the sunburst

- values : sets the values associated with sunburst sectors,determaining their width

```
import plotly.express as px
```

```
data = px.data.tips()
data.head()
```

```
px.sunburst(data,path=['sex','smoker'],values='total_bill')
```

```
px.sunburst(data,path=['day','smoker'],values='total_bill')
```

```
px.sunburst(data,path=['day','smoker','sex'],values='total_bill')
```

```
df = px.data.iris()
df.head()
```

data[['species', 'species_id']]: Selects the columns 'species' and 'species_id' from the DataFrame data.

.groupby('species'): Groups the resulting DataFrame by the values in the 'species' column. This means that rows with the same 'species' value are grouped together.

.min(): Applies the min() function to each group. This calculates the minimum value for each 'species_id' within each group of 'species'.

```
df[['species', 'species_id']].groupby('species').min()
```

```
px.parallel_coordinates(df,color="species_id")
```
