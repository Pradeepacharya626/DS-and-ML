# Funnel charts 

Funnel charts are a type of chart often used to represent stages in a process and show the amount of data that passes through each stage. They are particularly useful for visualizing processes that have sequential steps, such as sales pipelines, marketing conversion rates, or any process where data is progressively filtered down.

Funnel charts are often used to represent data in different stages of a buissness process. It is a important machanism in buissness intelligence to find the potential problem areas of process

```
import plotly.express as px
```

```
data = dict(
    number=[39, 25, 15, 10],
    stage=["Website visit", "Downloaded", "Installed", "Contacted"]
)
data
```

```
px.funnel(data, x='number', y='stage')
```

```
import pandas as pd

df_mtl = pd.DataFrame(dict(
    number=[39, 25, 15, 10],
    stage=["Website visit", "Downloaded", "Installed", "Contacted"]))
df_mtl['office'] = "Montreal"

df_toro = pd.DataFrame(dict(
    number=[52, 36, 22, 18],
    stage=["Website visit", "Downloaded", "Installed", "Contacted"]))
df_toro['office'] = "Toronto"

df = pd.concat([df_mtl, df_toro])
df
```

```
px.funnel(df, x='number', y='stage', color='office')
```

# gantt charts

A Gantt chart is a type of bar chart that represents a project schedule. It is named after its inventor, Henry Gantt, who designed it in the 1910s. Gantt charts are commonly used in project management to illustrate the start and finish dates of the various elements of a project. 

```
import plotly.figure_factory as ff
```

```
df = [dict(Task="Task #1", Start='2009-01-01', Finish='2009-02-28'),
      dict(Task="Task #2", Start='2009-03-05', Finish='2009-04-15'),
      dict(Task="Task #3", Start='2009-02-20', Finish='2009-05-30')]

fig = ff.create_gantt(df)
fig.show()
```
