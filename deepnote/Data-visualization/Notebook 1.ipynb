## Multivariate Analysis

### Heat Map

A heat map is a data visualization technique that shows the magnitude of a phenomenon as color in two dimensions. In a heat map, data values are represented by color, making it easy to see patterns, correlations, and anomalies. Heat maps are often used to visualize the correlation matrix of a dataset, where each cell in the matrix shows the correlation between two variables.


The range of -1 to 1 typically refers to the correlation coefficient in statistics. The correlation coefficient measures the strength and direction of a linear relationship between two variables. Hereâ€™s what the values indicate:

- **1**: Perfect positive correlation. As one variable increases, the other variable increases proportionally.
- **0**: No correlation. There is no linear relationship between the variables.
- **-1**: Perfect negative correlation. As one variable increases, the other variable decreases proportionally.

Values between -1 and 1 indicate the degree of correlation, with values closer to 1 or -1 indicating stronger correlations, and values closer to 0 indicating weaker correlations.

data_numeric.corr() computes the pairwise correlation of columns.

annot=True to annotate the cells with correlation coefficients and fmt='.1f' to format these coefficients to one decimal place.


```
# Before creating the heatmap, we need to ensure only numeric data is included
# Extracting numeric columns from the dataframe
data_numeric = data.select_dtypes(include=["number"])

plt.rcParams['figure.figsize'] = (19, 8)
sns.heatmap(data_numeric.corr(), annot=True, fmt='.1f')
plt.show()
```
